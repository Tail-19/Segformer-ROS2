```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("./frame0000.jpg",)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
tmp = np.ones((img.shape[0], img.shape[1])).astype(np.float32)
tmp[np.where((img[:,:,0]==0) & (img[:,:,1]==128) & (img[:,:,2]==1))]=0.05
img1 = cv2.resize(tmp, (128, 64))
# img1 = cv2.medianBlur(img1, 5)
# kernel = np.ones((17, 17), np.uint8)
# img1 = cv2.erode(img1, kernel) 
# img1[np.where(img1 == 0)] = 0.05
semantic_img = img1
```

```python
print(img[-1][-1])
plt.imshow(semantic_img)
```

```python
semantic_img = cv2.medianBlur(semantic_img, 5)
plt.imshow(semantic_img)
```

```python
kernel = np.ones((7, 7), np.uint8)
semantic_img_1 = cv2.dilate(semantic_img, kernel, iterations = 1) 
plt.imshow(semantic_img_1)
```
